```python
# Import all the libraries we will be using
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# read te data csv file
movies_df = pd.read_csv("movies.csv",encoding="ISO-8859-1")
movies_df.head()
```

```python
# Let's see how many null values do we have in %

for col in movies_df: # Go throught all columns
    pct_null = np.mean(movies_df[col].isnull()) # calculate how much null values there is ( in % )
    print(f"Column name : {col} \t==> Percent of null values = {pct_null} % ")
```

```python
# Replace null (NaN) values with 0
for col_name in movies_df: # go throught columns
    if movies_df[col_name].dtypes != 'object': # where column is int64 or float64
        movies_df [col_name] = movies_df [col_name].replace(np.nan, 0 )#Replace null values with 0 using numpy np.nan

movies_df
```

```python
# We should replace the released column int 2 columns ( realease_date & released_in).
# We will then extract the year from realease_date and compare it to the year column 
    # to see if there is any difference in the date
```

```python
# Get the released_date column
movies_df["released_date"] = movies_df["released"].str.split("(").str.get(0).str.strip()
```

```python
# extract the released year from released_date column
movies_df["year_released"] = pd.DatetimeIndex(movies_df.released_date).year
```

```python
# Get the released_in column
movies_df["released_in"] = movies_df["released"].str.split("(").str.get(1).str.strip().str.replace(')',' ')
```

```python
# Check the dataframe column types
movies_df.dtypes
```

```python
# We will try and change column type 

# Change budget data type from float to int
movies_df.budget = movies_df.budget.astype('int64') # Convert budget column to integer type

# Change gross data type from float to int
movies_df.gross = movies_df.gross.astype('int64')

# Change votes data type from float to int
movies_df.votes = movies_df.votes.astype('int64')

# Change runtime data type from float to int
movies_df.runtime = movies_df.runtime.astype('int64')

# Convert it from float to object 
movies_df.year_released = movies_df.year_released.astype('object')
```

```python
# there is lots of fdifferences between the year column and the year_released column data

movies_df[movies_df.year_released != movies_df.year]
```

```python
# Sort the data using gross ( revenue ) column in an ascending order
movies_df.sort_values( "gross", ascending=False).head()
```

```python
#Let's now try and find the correlations between those columns

movies_df.corr(method="pearson") 
# Methods we can use for correlation :
    # pearson by deafault
    # kendall
    # spearman
```

```python
# resizing the graphical output
plt.rcParams['figure.figsize'] = (12, 6)

my_correlation = movies_df.corr()
sns.heatmap(my_correlation, annot=True, linewidths=1) # show heatmap

plt.title("Correlation Matrix") # set title
plt.xlabel("Correlated columns") # set the x axis label
plt.ylabel("Correlated columns")

plt.show()
```
